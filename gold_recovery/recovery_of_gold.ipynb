### Проверка корректности расчета пaраметра recovery (эффективности обогащения руды) на примере обучающей выборки

Проверим корректность расчетов показателя эффективности обогащения руды на примере обучающей выборки. Для этого проведем расчет параметра recovery на основании формулы с учетом показателей доли золота в концентрате на входе (до флотации и до первой стадии оичтки), доли золота в руде на выходе (после флотации и полного цикла очистки) и доли золота в отвалах (после флотации и полного цикла очистки) и сравним полученный расчет со значениями, представленными в таблице, определим среднюю абсолютную ошибку для полученных данных. 

```python
df_train['rougher.output.recovery'] #recovery предоставленных данных после флотации
```

```python
#формула расчета recovey для проверки
recovery_check = (df_train['rougher.output.concentrate_au'] *(df_train['rougher.input.feed_au'] - df_train['rougher.output.tail_au']))/(df_train['rougher.input.feed_au'] * (df_train['rougher.output.concentrate_au'] - df_train['rougher.output.tail_au'])) * 100
```

```python
recovery_check
```

```python
MAE_recovery_check = (recovery_check - df_train['rougher.output.recovery']).mean() #средняя абсолютная ошибка расчета recovery
MAE_recovery_check
```

Низкое значение MAE полученных результатов (погрешность вычислений) говорит о корректности значений recovery в предложенных данных.

### Анализ изменения концентрации металлов на различных этапах очистки

Для анализа концентрации металлов (Au, Ag, Pb) будем исползовать средние значения и диаграммы распределение концентрации в отношении каждого этапа очистки.

```python
#функция исследования концентрации металлов
def metall_concentration_research(df):
    
    for i in ['au', 'ag', 'pb']: 
        conc_mean = []
        plt.figure(figsize=(15, 4))
        for k in ['rougher' , 'primary_cleaner', 'final']:
            conc_mean.append(df[f'{k}.output.concentrate_{i}'].mean())
            sns.distplot(df[f'{k}.output.concentrate_{i}'], kde=False, label= f'{k}')
        
        plt.grid()    
        plt.title(label=f'Distribution diagram {i}')
        plt.xlabel('concentration')
        plt.ylabel('frequency')
        plt.legend(loc='upper right')
        plt.show()
        
        stage = ['rougher', 'cleaner', 'final']
        plt.figure(figsize=(5, 6))
        plt.bar(stage, conc_mean)
        plt.title(label=f'Сoncentration on stages of cleaner {i}')
        plt.ylabel('mean_concentration')
        plt.grid(axis='y')
        plt.show()
        
        df_conc_mean = pd.DataFrame({'stage': stage, f'mean_concentration {i}': conc_mean})
        print(df_conc_mean)
    
```

```python
metall_concentration_research(df_full) #исследование концентраций металлов на исходной выборке
```

```python
metall_concentration_research(df_train) #исследование концентраций металлов на обучающей выборке
```

Как видно из поученных данных, динамика изменения значения концентрации для разных металлов различна. Для золота концентрация на каждом последующем этапе очистки повышается, что диктуется собственно технологической задачей производства, концентрация серебра же наоборот с каждым новым этапом снижается (что таже вероятно определяется задачами очистки: серебро менее ценный металл, а потеря его концентрации диктуется приоритетом выделения золота), для свинца же характерно увеличение коцентрации только после флотации, на последующих этапах очистки концентрация этого металла практически не меняется. Аналогичным образом ведут себя оси "колоколов" диаграмм нормального распределения концентраций на разных этапах, по сути - соответствую средним значениям. ###  Исследование суммарных концентраций всех веществ на разных стадиях производства.

Проведем исследование суммарной концентрации всех веществ на разных стадиях производства. Проведем суммирование по четырем составляющим концентрата продукта для каждой стадии производства, найдем распределения полученной суммарной величины и ее среднее значение для каждого из этапов технологического цикла.

```python
#функция исследования cуммарной концентрации всех веществ на различных этапах очистки
def all_sub_concentration_research(df):

    concentration_on_stages = []
    plt.figure(figsize=(16, 4))
    count_zero_sum_conc = []
    for k in ['rougher.input.feed_', 'rougher.output.concentrate_', 
              'primary_cleaner.output.concentrate_', 'final.output.concentrate_']:
        sum_conc = 0
        df_disp_sum_conc = pd.Series([0]*len(df['date']))
        for i in ['au', 'ag', 'pb', 'sol']:
            sum_conc += df[f'{k}{i}'].mean()
            df_disp_sum_conc += df[f'{k}{i}']
        
        sns.distplot(df_disp_sum_conc, kde=False, label= f'{k}')   
        
        concentration_on_stages.append(sum_conc)
        
        df[f'{k}sum'] = df_disp_sum_conc #добаление столбца суммарной концентрации веществ для этапа очистки
        count_zero_sum_conc.append(df[df[f'{k}sum']==0]['date'].count()) #количество значениний с 0-ой суммарной концентрацией
    
    plt.grid()   
    plt.title(label='Distribution diagram sum_concentration')
    plt.xlabel('sum_concentration')
    plt.ylabel('frequency')
    plt.legend(loc='upper left')
    plt.show()    
    
    stage = ['rougher.input', 'rougher.output', 'primary_cleaner.output', 'final.output']
    plt.figure(figsize=(8, 4))
    plt.bar(stage, concentration_on_stages)
    plt.title(label=f'Sum concentration on stages of cleanning')
    plt.ylabel('mean_sum_concentration')
    plt.grid(axis='y')
    plt.show()
    
    df_conc_sum_mean = pd.DataFrame({'stage': stage, f'mean_sum_concentration': concentration_on_stages,
                                    'count_zero_sum_conc':count_zero_sum_conc})
    print(df_conc_sum_mean)
```

###  Подготовка функции вычисления метрики sMAPE

```python
#подготовка функции вычисления метрики sMAPE

def smape(y, y_pred, **kwargs):
    
    smape = (abs(y - y_pred)/((abs(y) + abs(y_pred))/2)).mean()*100

    return smape
```

```python
sMAPE = make_scorer(smape, greater_is_better=False) #создание оценивающей метрики для процедуры кросс-валлидации
```

###  Выбор наилучшего сочетания моделей с учетом оптимизации гипперпараметров для минимизации метрики sMAPE

Создадим функцию, которая на основе поданной целевой выборки (данные этапа флотации или данные этапов очистки) определит оптимальную модель обучения (линейная регрессия, решаюшее дерево или случайный лес) и ее гипперпараметры на основании минимизации рассматриваемой метрики sMAPE. Подбор гиперпараметров будем производить перебором в цикле (циклах) для более точной настройки модели. 

```python
def best_smape_meth(target):
    #проверка линейной регрессии
    smape_lr = cross_val_score(LinearRegression(), features_train_scaled, target, scoring=sMAPE, cv=5).mean()
    print('smape_lr: ', smape_lr)
    print()
    
    #проверка решающего дерева 
    best_smape_dt = -1000
    best_depht_dt = 0
    for depth in range(2, 31, 2):
        smape_dt = cross_val_score(DecisionTreeRegressor(random_state=12345, max_depth=depth), 
                                   features_train_scaled, target, scoring=sMAPE, cv=5).mean()
        if smape_dt > best_smape_dt:
            best_smape_dt = smape_dt
            best_depht_dt = depth
        print('depht_dt:', depth, '; smape_dt for depht_dt:', smape_dt, '; best_smape_dt:', best_smape_dt)
    print()
            
    #проверка случайного леса 
    best_smape_rf = -1000
    best_est_rf = 0
    best_depth_rf = 0

    for est in [1, 3 , 5 , 7 , 9, 25, 50]:
        for depth in range(1, 12, 2):
            smape_rf = cross_val_score(RandomForestRegressor(random_state=12345, n_estimators=est, max_depth=depth), 
                                       features_train_scaled, target, scoring=sMAPE, cv=5).mean()
            if smape_rf > best_smape_rf:
                best_smape_rf = smape_rf
                best_est_rf = est
                best_depth_rf = depth
            print('n_estimators_rf:', est, '; depht_rf:', depth, '; smape_rf:', smape_rf, '; best_smape_rf:', best_smape_rf)
    print()
    
    # блок определения лучших параметров и вывод результатов
    best_smape = max(smape_lr, best_smape_dt, best_smape_rf)
    
    if best_smape == smape_lr:
        print('Best model is LinearRegression: sMAPE=', smape_lr)
    elif best_smape == best_smape_dt:
        print('Best model is DecisionTreeRegressor: sMAPE=', best_smape_dt, '; best_depth=', best_depht_dt)
    else:
        print('Best model is RandomForestRegressor: sMAPE=', best_smape_rf, '; best_n_estimators=', best_est_rf,
              '; best_depth=', best_depth_rf)
    
    return best_smape
```

```python
best_smape_rougher = best_smape_meth(target_train_rougher)
best_smape_rougher
```

```python
best_smape_final = best_smape_meth(target_train_final)
best_smape_final
```

 - Проведен исследовательский анализ данных, включающий: анализ изменения концентрации металлов на различных стадиях очистки (для золота концентрация на каждом последующем этапе очистки повышается, для серебра с каждым новым этапом концентрация снижается, для свинца характерно увеличение коцентрации только после флотации, на последующих этапах очистки концентрация этого металла практически не меняется), сравнение распределения размеров гранул исходного сырья в обучающей и тестовой выборках (распределение идентично), исследование суммарных концентраций всех веществ на разных этапа очистки (динамика носит нелинейный характер).
 - Проведены подготовка и выбор модели машинного обучения . Принято решение о разделении обучения на прогнозирование на этапе флотации и на прогнозирование на финальном этапе очиски с нахождением самостоятельных моделей по каждому указанному этапу. Проведены подготовка признаков и стандартизация выборок. Создана функция вычисления метрики sMAPE. В качестве наилучших моделей признаны модели "случайного леса" для этапа флотации и финальной очистки с определенными гипперпараметрами (глубиной и количеством деревье): 3 и 3 - для флотации; 5 и 25 - для финальной очистки соответсвенно.
 - Проверка модели на тестовой выборке.
